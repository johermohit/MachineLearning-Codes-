What we have to do in this project?
First of all we have to Generate 1000 random paragraphs of random lengths with random number of words and random number of spaces. The maximum length of each paragraph can be 250 characters including spaces while the minimum ids 75 characters.Whreas the minimum length of each word is 2 and maximum is 10 and each paragraph will be comprised only of lower case charcters.
the next thing we have to do is convert each paragraph into a D Dimensional vector.
Now we Have to generate an another paragraph also satisfying the above conditions and we have to check that randomly generated paragraph is similiar to which one of the paragraphs.


```python
# Here we import random and numpy
import random
import numpy as np
```

```python
#Now we do is to generate 1000 paragraphs 
alphabet='abcdefghijklmnopqrstuvwxyz'
paragraphs=[]
for i in range(1000):
    paralength=random.randint(75,250)(#Here we randomly generate the paragraph length)
    string=[]
    while(True):
        s=''
        for i in range(random.randint(2,10)):(#here we randomly generate the word size)
            r1=random.randint(0,25)
            s+=alphabet[r1]
        string.append(s+ " ")
        paragraph="".join(string)
        if len(paragraph) >=(paralength-10):
            break
    paragraphs.append(paragraph)(#here we append all the paragraphs into a list)
```

```python
#Here we split all the words from each paragraph and append it in a list whis consists of 1000 lists
wordlist=[]
for i in paragraphs:
    wordlist.append(i.split(" ")) 
#Here we find the unique word from that lists
uniqueword=[]
for i in wordlist:
    uniquewor,count=np.unique(i,return_counts=True)
    uniqueword.append(dict(zip(uniquewor,count)))
```

```python
#Here we find the frequency of all unique word from each paragraph contained in a list
frequency=[]
for i in uniqueword:
    fre=[]
    for key in i:
         fre.append(i.get(key))
    frequency.append(fre)
frequency_array=np.array(frequency)
```

Now we have to convert each paragraph into a D dimensional vector 
the first things come in our mind is what is D dimensional vector?
we can say it in simple words that we have to convert each paragraph into a one dimensional numpy array of D integer numbers.
then what will be the value of D ?
D will be equal to the number of unique words in a paragraph. And these integer numbers is called word frequencies

```python
D_vector=np.array(frequency_array)
#Here  it makes the numpy array of all D dimensional vector
```
