```python
def RrlaighPrameter(m,RVmin,binsize,RVno):
    global Mode
    global MiniMumValueOfRV 
    global TotalBins
    global TotalNumberofValuesofRV
    global Probablity
    Mode = m
    MiniMumValueOfRV =RVmin
    TotalBins =binsize
    TotalNumberofValuesofRV =RVno
```

```python
def PotRailagh():
    global X
    global Z
    global Y
    
    global Probablity

    RVValues = []
    Probablity=[]
    l = ((Mode)**2)
    m = math.exp(-(((MiniMumValueOfRV)**2)/(2*l)))

    for i in range(0,TotalNumberofValuesofRV):

        y = random.uniform(0,1)
        Probablity.append(y)

        n = math.log((1/(m - y))**2)

        K = math.sqrt((l * n))

        RVValues.append(K)

    MaximumValueRV = max(RVValues)

    MaximumValueRV = math.ceil(MaximumValueRV)

    BinWidth = (MaximumValueRV - MiniMumValueOfRV)/TotalBins

    Frequency = [0] * np.linspace(MiniMumValueOfRV,MaximumValueRV,TotalBins)

    for i in range(0,TotalNumberofValuesofRV):

        Frequency[math.floor((RVValues[i]-MiniMumValueOfRV)/BinWidth)] += 1
    
    XAxisValues = [1] * np.linspace(MiniMumValueOfRV,MaximumValueRV,TotalBins)
    a=(Frequency/TotalNumberofValuesofRV)
    plt.bar(XAxisValues,a,BinWidth)
    temp=RVValues
    Z=XAxisValues
    
    Mode2=165
    MiniMumValueOfRV2=40
    TotalNumberofValuesofRV2=30
    RVValues2 = []
    l = ((Mode2)**2)
    m = math.exp(-(((MiniMumValueOfRV2)**2)/(2*l)))

    for i in Probablity:


        n = math.log((1/(m - i))**2)

        K = math.sqrt((l * n))

        RVValues2.append(K)

    MaximumValueRV2 = max(RVValues)

    MaximumValueRV2 = math.ceil(MaximumValueRV)

    BinWidth = (MaximumValueRV2 - MiniMumValueOfRV2)/TotalBins

    Frequency = [0] * np.linspace(MiniMumValueOfRV2,MaximumValueRV2,TotalBins)

    for i in range(0,TotalNumberofValuesofRV2):

        Frequency[math.floor((RVValues2[i]-MiniMumValueOfRV2)/BinWidth)] += 1
    
    XAxisValues = [1] * np.linspace(MiniMumValueOfRV2,MaximumValueRV2,TotalBins)
    a=(Frequency/TotalNumberofValuesofRV2)
    
    plt.bar(XAxisValues,a,BinWidth,edgecolor='red')
    Z=Probablity
    X=RVValues 
    Y=RVValues2
    return(plt.show())
    
```
